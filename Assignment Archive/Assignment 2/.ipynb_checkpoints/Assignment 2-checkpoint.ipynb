```python
# (a) Write your own kNN classifier

import numpy as np

class Knn:
# k-Nearest Neighbor class object for classification training and testing
    def __init__(self):
        self.x_train = None
        self.y_train = None
        
    def fit(self, x, y):
        # Save the training data to properties of this class
        # this would train our classifier, or in this case just memorizing the training data
        # x is the training data
        # y is the training labels where y[i] is the label for x[i]
        self.x_train = x
        self.y_train = y
        
#use python modules to calculate distance        
    def predict_builtin(self, x, k):
        y_hat = [] # Variable to store the estimated class label for 
        # Calculate the distance from each vector in x to the training data
        for i in range (x.shape[0]):
            tmp = np.array(x.iloc[i,:]).reshape(1,-1)
            dist = cdist(self.x_train, tmp, metric="euclidean")
            #make so can find most frequent class and nearest k value
            y_hat.append(self.y_train.iloc[dist.argsort(axis=0)[:k].flatten(),0].mode()[0])
        return y_hat
    
#calculate distance by hand
    def predict(self,x,k):
        y_hat = [] # Variable to store the estimated class label for 
        # Calculate the distance 
        for i in range (x.shape[0]):
            tmp = np.array(x.iloc[i,:])
            dist = np.tile(tmp, x.shape[0]).reshape([x.shape[0],x.shape[1]])
            dist = (dist - self.x_train)**2
            dist = (np.sum(dist, axis=1))**0.5
            #make so can find most frequent class and nearest k value
            y_hat.append(self.y_train.iloc[dist.argsort(axis = 0)[:k],0].mode()[0])
        return y_hat
    
# Metric of overall classification accuracy
#  (a more general function, sklearn.metrics.accuracy_score, is also available)
def accuracy(y,y_hat):
    nvalues = len(y)
    #accuracy = sum(y == y_hat) / nvalues
    num = 0
    for i in range(len(y_hat)):
        if y_hat[i] == y[i]:
            num += 1
    accuracy = num / nvalues
    return accuracy
    
```

```python
x_train = ([3.19,9.26,9.38,8.77,7.91,3.79,3.18,7.61,2.36,6.26,6.62,1.53,6.25,7.93,7.07,
           4.58,4.14,2.14,9.04,4.56,3.99,6.71,2.51,0.84,6.13,5.22,0.25,3.60,1.36,5.59,
           4.81,1.14,0.36,2.31,1.37,5.86,4.23,9.48,2.26,0.77,4.33])
y_train = ([46.40,172.16,209.00,203.31,82.88,62.57,14.38,177.00,8.01,82.35,84.84,-5.59,
           54.96,167.17,83.09,-21.63,94.64,63.97,106.57,38.99,88.26,66.99,-11.12,-0.44,
           65.08,61.47,-0.61,23.85,10.55,83.69,54.35,51.82,-18.63,1.98,4.90,55.44,50.09,
           155.66,45.57,18.12,30.58])

x_test = ([5.65,0.07,8.84,5.14,6.65,1.35,5.45,7.39,3.35])
y_test = ([98.52,16.09,198.45,75.90,85.11,47.64,14.76,141.03,-39.13])
```

```python
#a
#create scatter plot of training data
plt.title("Training Data")
plt.scatter(x_train, y_train)
plt.xlabel('x')
plt.ylabel('y')
```

```python
#b 

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import itertools

#reshape data so can be used in LinearRegression
x_train = np.array(x_train).reshape(-1,1)
y_train = np.array(y_train).reshape(-1,1)

x_test = np.array(x_test).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)

#linear model
model1 = LinearRegression().fit(x_train, y_train)
print('Linear Relationship - Linear Regression Estimate: y = ',
     round(model1.intercept_[0],3), '+',
     round(model1.coef_[0][0],3), 'x')

#r squared and MSE
pred1 = model1.predict(x_train)
ssreg = np.sum((y_train - pred1) * (y_train - pred1))
sse = np.sum((y_train - np.mean(y_train)) * (y_train - np.mean(y_train)))

rsq1 = 1 - (ssreg / sse)
mse1 = ssreg / len(y_train)
print('Linear Relationship R-Squared: ', round(rsq1,3))
print('Linear Relationship MSE: ', round(mse1, 3))
```

```python
#c
#create exponential x term
thirty_exp = 30**x_train
x_exp_train = np.column_stack((x_train, thirty_exp))

#refit linear model
model2 = LinearRegression().fit(x_exp_train, y_train)
print('Non-Linear Relationship - Linear Regression Estimate: y = ',
     round(model2.intercept_[0],3), '+',
     round(model2.coef_[0][0],3), 'x')

#r squared and MSE
pred2 = model2.predict(x_exp_train)
ssreg2 = np.sum((y_train - pred2) * (y_train - pred2))
sse2 = np.sum((y_train - np.mean(y_train)) * (y_train - np.mean(y_train)))

rsq2 = 1 - (ssreg2 / sse2)
mse2 = ssreg2 / len(y_train)
print('Non-Linear Relationship R-Squared: ', round(rsq2,3))
print('Non-Linear Relationship MSE: ', round(mse2, 3))
```

```python
#d
#prediction interval
interval = np.arange(min(x_train), max(x_train), step = .01)
interval = np.array([interval]).reshape(-1,1)
interval2 = interval*interval
interval2 = np.column_stack((interval, interval2))

#set up predictions
y_hat1 = model1.predict(interval)
y_hat2 = model2.predict(interval2)

#plot predictions and data
plt.scatter(x_train, y_train, label = 'Original Data')
plt.plot(interval, y_hat1, 'r', label = 'Linear relationship')
plt.plot(interval, y_hat2, 'g', label = 'Nonlinear relationship')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression on Training Data')
plt.legend()
plt.show()
```

```python
#e
#apply models to test data
#linear relationship
pred3 = model1.predict(x_test)
rsq3 = r2_score(y_test, pred3)
mse3 = mean_squared_error(y_test, pred3)
print('Test Data - Linear Relationship R-Squared: ', round(rsq3,3))
print('Test Data - Linear Relationship MSE: ', round(mse3, 3))


#nonlinear relationship
thirty_exp2 = 30**x_test
x_exp_test = np.column_stack((x_test, thirty_exp2))
pred4 = model2.predict(x_exp_test)
rsq4 = r2_score(y_test, pred4)
mse4 = mean_squared_error(y_test, pred4)
print('Test Data - Non-Linear Relationship R-Squared: ', round(rsq4,3))
print('Test Data - Non-Linear Relationship MSE: ', round(mse4, 3))


```
