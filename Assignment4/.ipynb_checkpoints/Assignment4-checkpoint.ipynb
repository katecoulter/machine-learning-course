```python
# a 
# implement k means algorithm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

class kmeans:    
    def __init__(self, k_clusters=5, random_state=None, tol=.0001, metric="l2"):
        #initialize function
        self.k_clusters = k_clusters
        self.random_state = random_state
        self.tol = tol
        self.metric = metric
    
    #fit to k means algorithm
    def fit(self, X):        
        n, p = X.shape
        if X.shape[0] < self.k_clusters:
            self.k_clusters = X.shape[0]
        if self.random_state != None:
            np.random.seed(self.random_state)
        self.c = X[np.random.choice(n, self.k_clusters, replace=False)]
        self.labels = np.zeros(X.shape[0], dtype=int) - 1
        diff = 1
        
        #set so that will continue to loop unless distance is less than tol
        while self.tol < diff:
            self.labels = self.cal_labels(X)
            next_centroid = self.cal_centroid(X)
            diff = np.max(np.linalg.norm(next_centroid - self.c, axis=1))
            self.c = next_centroid
        return self
    
    #determine class
    def cal_labels(self, X):
        labels = np.zeros(X.shape[0], dtype=int)
        for i, point in enumerate(X):
            labels[i] = np.argmin(np.linalg.norm(self.c - point, axis=1))
        return labels
    
    #calculate centroid 
    def cal_centroid(self, X):
        centroid = np.zeros_like(self.c)
        for i in range(centroid.shape[0]):
            centroid[i] = np.mean(X[np.where(self.labels == i)], axis=0)
        return centroid
    #for elbow part later
    def elbow(self, X):
        elbow_list = np.zeros(X.shape[0])
        for i, point in enumerate(X):
            elbow_list[i] = np.linalg.norm(point - self.c[self.labels[i]])
        return np.sum(elbow_list ** 2)
```

```python
#function to make plot
def getplot(data, labels, figsize=(9,6), title=None, xlab=None, ylab=None, legend=True):
    plt.figure(figsize=figsize)
    for i in np.unique(labels):
        plt.scatter(data[np.where(labels == i), 0], data[np.where(labels == i), 1], 
                    cmap=plt.cm.get_cmap('Spectral', 10), label=i)
    if legend:
        plt.legend()
    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.show()
```

```python
#create and implement on two datasets 
from sklearn.datasets import make_blobs

# 2 centers
x2, y = make_blobs(n_samples=500, n_features=2, centers=2, random_state=100)
getplot(x2, y, figsize=(9, 7), title="Created Blob Dataset with Two Centers", xlab="x1", ylab="x2")

SSE = []
for k in range(1, 11):
    kmean = kmeans(k_clusters=k, random_state=0).fit(x2)
    SSE.append(kmean.elbow(x2))
plt.figure(figsize=(9,7))
plt.plot(np.arange(1, 11), SSE, '-o')
plt.title("Elbow Curve of SSE and Two Center Blob Data")
plt.xlabel("Number of Clusters")
plt.ylabel("Sum of Square Error")
plt.show()
```

For our dataset that has 2 centers, the "elbow" appears to be a $k = 2$. Prior to this $k$ value, the SSE has a very large value. However, once we increase $k$ beyond a value of 2, the SSE remains near constant with just a small decrease as $k$ increases. If we look at our original plot, our data is clearly a perfect case where we can cluster our dataset into 2 groups, which implies that a $k$ value greater than 2 would overfit the data.

```python
# 5 centers
x5, y = make_blobs(n_samples=500, n_features=2, centers=5, random_state=100)
getplot(x5, y, figsize=(9, 7), title="Created Blob Dataset with Five Centers", xlab="x1", ylab="x2")

SSE = []
for k in range(1, 11):
    kmean = kmeans(k_clusters=k, random_state=0).fit(x5)
    SSE.append(kmean.elbow(x5))
plt.figure(figsize=(9,7))
plt.plot(np.arange(1, 11), SSE, '-o')
plt.title("Elbow Curve of SSE and Five Center Blob Data")
plt.xlabel("Number of Clusters")
plt.ylabel("Sum of Square Error")
plt.show()
```

For our dataset that has 5 centers, the "elbow" appears to be a $k = 4$. Prior to this $k$ value, the SSE has very large values and each increase in $k$ results in a dramatic decrease in SSE. However, once we increase $k$ beyond a value of 4, the SSE remains near constant for values of $k = 5$ and $k = 6$, with an additional small decrease in SSE as $k = 7$. If we look at our original data, we can see some overlap in the groups, specifically groups 1 and 3. What our Elbow Curve seems to indicate is that our data naturally seems to fall into 4 clusters, despite us knowing that it was created with 5 centers. 

(b) Clustering techniques:

- __K-means__: This clustering technique is a hard clustering method which assigns $n$ data objects to a pre-defined number of exactly $k$ clusters. Membership is determined by calculating the centroid for each group and then optimizing the cost function defined by the sum of squares.

- __Agglomerative Clustering__: A type of hierarchical clustering technique that uses a dendrogram built from the bottom level by merging the most similar pair of clusters and stopping only when all data points are merged into a single cluster. 

- __Gaussian Mixture Models__: Provides flexibility by assuming the data points are Gaussian distributed, thus providing two parameters that can be used to describe the shape of the cluster - the mean and the standard deviation. To find the parameters for each cluster, you use the optimization algorithm called Expectation-Maximization (EM).

- __DBSCAN__: As this clustering method does not assume a shape or amount of clusters in the dataset, DBSCAN identifies high density, at times arbitrarily shaped clusters while also being able to identify outliers as noise. Does not perform well with clusters of varying densities or with very high dimensional data.

- __Spectral Clustering__: Performs dimensionality reduction prior to clustering the data by using the spectrum (eigenvalues) of the similarity matrix of the data. This technique clusters data that may not necesarrily be compact or within the same convex boundaries.

```python
# a
# load the scikit-larn digits dataset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import matplotlib.style as style 
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn import datasets

digits = datasets.load_digits()
x = digits.data
y = digits.target

# apply PCA and reduce the data into a 2D space
# note that the following code relies on the chapter in "Python Data Science Handbook"
# available at: https://jakevdp.github.io/PythonDataScienceHandbook/05.09-principal-component-analysis.html

print(digits.data.shape)
# shape is 1797, 64
# so project from 64 to 2D
pca = PCA(2)
projected = pca.fit_transform(digits.data)
print(projected.shape)
# now in 2 dimensional space

# plot the resulting 2D data
plt.figure(figsize=(9,6))
style.use('ggplot')
sns.set_context('notebook', font_scale=1.2)
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=.7,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.title('PCA of MNIST Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar();
```

(a) continued -

Above is the plot of the 2-dimensional representation of the MNIST dataset using PCA. As you can see, there is overlap between clusters so the regions do not appear to be well defined.

```python
# b
# apply t-SNE to the digits dataset
import numpy as np
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(digits.data)

# plot in 2D
plt.figure(figsize=(9,6))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1],
            c=digits.target, edgecolor='none', alpha=0.7,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.title('t-SNE of MNIST Dataset')
plt.xlabel('x-tnse')
plt.ylabel('y-tnse')
plt.colorbar();
```

(b) continued - 

Above is the plot of the 2-dimensional representation of the MNIST dataset using t-SNE. The clusters here are far more defined than they were in the previous plot which used PCA.

(c) Both PCA and t-SNE are techniques used to reduce the dimensions of data, thereby making it less complex so that it is easier to work with, for example visualization is now possible and overfitting is less likely to occur. As is evident in our plots, t-SNE seems to cluster our data better than PCA. As mentioned, after performing PCA the clusters had quite a bit of overlap and did not seem to have well defined regions. In comparison, the plot that depicts our application of t-SNE on our data has distinct regions with hardly any overlap, and the overlap that does exist (between clusters 9 and 3, for example) appears to be the result of outliers in our data.

While t-SNE performed better in this case, there are benefits to using PCA instead. PCA results are deterministic, meaning that there is an easily interpretable solution that is a guaranteed solution, while t-SNE can result in multiple solutions, making it difficult to interpret at times. Additionally, new data can be projected using PCA while t-SNE has to approximate the parametric algorithm. The way that PCA is able to project new data, though, is by preserving global properties (like high variance) so that eigenvectors can be used to find new basis vectors.  Finally, t-SNE is much more computationally difficult than PCA.

```python
# extract all 5's from the digits dataset
import numpy as np
import matplotlib.pyplot as plt
from sklearn import decomposition
from sklearn import datasets

digits = datasets.load_digits()
x = digits.data
y = digits.target

fives = digits.data[digits.target == 5,:]
```

```python
# a
# plot examples of the original images
r = np.random.choice(fives.shape[0], 15, replace=False)
fig, axs = plt.subplots(3, 5, figsize=(9,5))
for ax, k in zip(axs.flatten(), r):
    ax.imshow(fives[k].reshape(8,8), cmap='gray')
    ax.set_xticks([])
    ax.set_yticks([])
plt.suptitle('15 Sample Images of Fives',y=1.05, fontsize=20)
#plt.tight_layout()
plt.show()
```

```python
# b
# perform PCA on the data
n = 64
exp_var = []
for n in range(1, n+1):
    pca = PCA(n)
    projected = pca.fit_transform(x)
    exp_var.append(sum(pca.explained_variance_ratio_))
    pass

plt.figure(figsize=(9,6))
plt.plot(np.arange(n), np.array(exp_var), '-bo')
plt.title('Fraction of Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

(b) continued -

Note that this curve quantifies how much of the total, 64-dimensional variance is contained within the first $n$ components. For example, the first 10 components contain approximately 85% of the variance, and we need around 40 components to describe close to 100% of the variance.

```python
# c
# select an image from dataset
# will select randomly
from random import randrange
np.random.seed(123)
image = np.random.randint(fives.shape[0])

# plot the original image and the compressed versions 
# with k levels of compression = 1, 5, 10, 25
fig, axs = plt.subplots(1,5, figsize=(8,2))
for ax, k in zip(axs.flatten(), [0, 1, 5, 10, 25]):
    if k != 0:
        pca5 = PCA(n_components=k)
        fivespca = pca5.fit_transform(fives)
        fivespca_inv = pca5.inverse_transform(fivespca)
        im = fivespca_inv[image]
        title = "$k=%d$" % k
    elif k == 0:
        im = fives[image]
        title = "Original"

    ax.imshow(im.reshape(8,8), cmap="gray")
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])
plt.suptitle("Different Levels of Compression for Fives Image %d" % image, 
             y = 1.05, fontsize = 20)
plt.tight_layout()
plt.show()
```

(d) To well approximate the data in (c), the number of principal components depends on the particular case and reasoning for deciding to compress. If the goal is to keep the image as similar as possible, then a small compression rate such as $k = 25$ components should be chosen. However, if there are concerns about memory space, then a higher compression rate and smaller component value should be chosen, such as $k = 5$.

Compression achieved in each case:

For $k = 1$, the compression rate is found as $\frac{D-k}{D} = 98.44\%$. This level of compression is the blurriest of all our cases, and the color within the image has a lot of contrast, not allowing pixels to have a more gray color.

For $k = 5$, the compression rate is found as $\frac{D-k}{D} = 92.19\%$. This level of compression begins to allow different shades of gray, giving a bit more detail as to what the image may actually look like.

For $k = 10$, the compression rate is found as $\frac{D-k}{D} = 84.38\%$. This image starts to show the thinness of the writing, as it appeared quite thick and blurred before.

For $k = 25$, the compression rate is found as $\frac{D-k}{D} = 60.94\%$. This level is not quite as clear as the original as it still appears lighter in the background than the sharp black in the original image, but it looks quite similar and is a good approximation as it is a very similar image.
